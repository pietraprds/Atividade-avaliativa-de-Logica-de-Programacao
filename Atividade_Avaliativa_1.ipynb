**Questão** 1 — Classificação de
Temperaturas (Condicionais + Loop)


```python
temperaturas = [18, 22, 30, 35, 28, 15, 40]
contador_frio = 0
contador_agradavel = 0
contador_quente = 0
for num in temperaturas:
  if (num < 20):
    print("Frio")
    contador_frio= contador_frio + 1

  elif (num >= 20 and num <= 30):
    print("Agradável")
    contador_agradavel = contador_agradavel + 1

  else:
    print("Quente")
    contador_quente = contador_quente + 1
```

**Questão** 2 — Sistema de Avaliação de
Alunos

```python
notas_finais = [4.5, 6.0, 7.8, 9.0, 5.5, 8.2]
contador_aprovado = 0
for notas in notas_finais:
  if (notas < 5):
    print("Reprovado")
  elif (notas >= 5 and notas <= 7):
    print("Recuperação")
  else:
    print("Aprovado")
    contador_aprovado = contador_aprovado + 1

total_alunos = len(notas_finais)
soma = contador_aprovado
porcentagem = (contador_aprovado / total_alunos) * 100

print(f"Quantidade de aprovados:{soma}")
print(f"Porcentagem de aprovados: {porcentagem:.2f}%")
```

Questão 3 — Monitoramento de
Consumo de Energia

```python
consumo_energia = [120, 95, 200, 180, 75, 220, 160]
soma =sum(consumo_energia)
media = soma/len(consumo_energia)
elementos_baixo = []
elementos_moderado = []
elementos_alto = []
contador_alto = 0
for num in consumo_energia:
 if (num < 100):
  elementos_baixo.append(num)

 elif (num >= 100 and num <= 180):
  elementos_moderado.append(num)

 else:
  elementos_alto.append(num)
  contador_alto = contador_alto

print(f"Baixo{elementos_baixo}")
print(f"Moderado{elementos_moderado}")
print(f"Alto{elementos_alto}")



print(f"-------------------------------------")
print(f"Total de consumo da semana {soma}")
print(f"Media semanal {media}")
if contador_alto >= 2:
  print(f"ALERTA: {contador_alto} dias tiveram preços altos")
print(f"-------------------------------------")
```

**Questão** 4 — Simulação de Carrinho de
Compras

```python
precos =  [50, 120, 30, 200, 80, 15]
lst_final = []
for num in precos:
 if(num < 50):
  #print(f"num*0.5")
  p = num * 0.95
 elif(num>= 50 and num <= 150):
  #print(f"num*0.10")
  p = num  * 0.90
 else:
  #print(f"num*0.15")
  p = num  * 0.85
 lst_final.append(p)

preco_total = sum(lst_final)
print(f"Preço total é igual a {preco_total}")
print(f"A lista de preços é {lst_final}")
```
